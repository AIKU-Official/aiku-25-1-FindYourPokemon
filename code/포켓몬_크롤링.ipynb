```python
import requests
from bs4 import BeautifulSoup
import json
import time
```

```python
# 중복 제거용 번호 저장 세트
seen_numbers = set()
results = []

def get_pokemon_info(poke_id):
    url = f"https://pokemonkorea.co.kr/pokedex/view/{poke_id}"
    res = requests.get(url)
    if res.status_code != 200:
        print(f"[{poke_id}] 실패")
        return None

    soup = BeautifulSoup(res.text, 'html.parser')

    try:
        number = soup.select_one('p.font-lato').text.strip().replace("No.", "").strip()

        # 중복 번호면 건너뛰기
        if number in seen_numbers:
            return None
        seen_numbers.add(number)

        name_tag = soup.select_one('h3')
        name_lines = name_tag.get_text(separator="\n").split("\n")
        name = name_lines[1].strip() if len(name_lines) > 1 else "이름 없음"

        types = [t.text.strip() for t in soup.select('span.img-type p')]

        category_tag = soup.find('h4', text='분류')
        category = category_tag.find_next('p').text.strip() if category_tag else "분류 없음"

        return {
            "id": number,
            "name": name,
            "types": types,
            "classification": category
        }

        print(f"번호: {number}")
        print(f"이름: {name}")
        print(f"타입: {', '.join(types)}")
        print(f"분류: {category}\n")

    except Exception as e:
        print(f"[{poke_id}] 파싱 오류: {e}")
        return None
```

```python
# ✅ 예시: 1~240번까지 시도
for poke_id in range(1, 240):
    info = get_pokemon_info(poke_id)
    if info:
        results.append(info)
    time.sleep(0.3)
```
